# Relativistic Doppler Shift

## The Doppler Factor
Consider a source of light moving toward an observer with speed $v$ at an angle $\theta$ to our line of sight:

![Rel_Dop_Shift](Figures/Rel_Doppler_Shift.png)

where $p_1$ is a photon emitted at time $t_1$ in the direction of Earth, and  $p_2$ is a photon emitted at time $t_2$ towards Earth. What is the time interval between the arrival of the two photons at the observer?

Going back to the Lorentz transformation for time:

$$
    t_2-t_1 = \gamma \left[(t'_2-t'_1)+\frac{u (x'_2-x'_1)}{c^2} \right]
$$

Here, the positive x direction is away from the observer, and the relevant velocity component that we need is 
$$
    u=-v\cos \theta
$$
where the minus sign arrises are the source is moving in the negative x direction. Also, the distance travelled by $p_1$ in the time interval $\Delta t'$ in the primed frame is simply
$$
    x'_2-x'_1 = \Delta x' = c \Delta t'.
$$
So, combining all of these together, the above expression becomes
$$
    \Delta t_{\rm obs} = \gamma \left(\Delta t' -\frac{v \cos \theta \Delta t' }{c} \right)
$$
Now we can simplify this a bit more to get
$$
    \Delta t_{\rm obs} = \gamma \Delta t' \left(1 -\frac{v \cos \theta}{c} \right) \equiv \frac{1}{D} \Delta t'
$$
where
$$
    D = \frac{1}{\gamma \left(1-\frac{v}{c}\cos \theta \right)}
$$
and $\theta = 0$ corresponds to motion directly towards the observer. $D$ is known as the *Doppler Factor*. It plays a role analogous to the Lorentz factor, but allows for motion at some angle to the line of sight. 

When $\theta=90^{\rm o}$, then $D=\frac{1}{\gamma}$ and $\Delta t = \gamma \Delta t'$

We can gain insight into the expression for $\Delta t$ by considering the above plot with a few additions

![Rel_Dop_Shift](Figures/Rel_Doppler_Shift_edit.png)

where $\Delta t$ is the time between emission of the photons. The distance between photons $p_1$ and $p_2$ in this setup is:
$$
    \Delta x = c \Delta t - v \cos \theta \Delta t
$$
And so the time mesaured between the arrival of the photons is
$$
    \Delta t_{\rm obs} = \frac{\Delta x}{c} = \Delta t \left(1 - \frac{v}{c} \cos \theta \right)
$$
Now, the emission of the photons happens at the same place in the rest (primed) frame of the emitting object - which means that $\Delta x'=0$, so $\Delta = \gamma \Delta t'$, and so we arrive at 

$$
    \Delta t_{\rm obs} = \gamma \Delta t' \left(1 - \frac{v}{c} \cos \theta \right)
$$
as we obtained previously. We see that the origin of the factor $\left(1 - \frac{v}{c} \cos \theta \right)$ is the motion of the source towards/aways from the observer.

## Time interval durations
Is the observed time interval shorter or longer than the time interval in the rest frame of the source ($\Delta t'=\Delta t_{\rm rest}$)?
Starting with:
$$
    \Delta t_{\rm obs} = \gamma \Delta t_{\rm rest} \left(1 - \frac{v}{c} \cos \theta \right)
$$

- We know that $\gamma > 1$ always.
- $\left(1 - \frac{v}{c} \cos \theta \right) <1$ when $\theta \leq 90$
- $\left(1 - \frac{v}{c} \cos \theta \right) >1$ when $90 < \theta \leq 180$

Thus the two factors act in the same sense when $90 < \theta \leq 180$ (that is, they are both $>1$), so when the source is receding
$$
    \Delta t_{\rm obs} > \Delta t_{\rm rest}
$$
which means the time interval in the rest frame is shorter than observed. Now, when $\theta<<90$ such that $\cos \theta$ is very large, then
$$
    \gamma \left(1 - \frac{v}{c} \cos \theta \right) < 1
$$
the observed time interval will be shorter than the rest frame interval
$$
    \Delta t_{\rm obs} < \Delta t_{\rm rest}.
$$
This can be very useful when you record an evenet and know the sources motion relative to you, as it can allow you to figure out how long it took in the rest frame of the source.

## Relativistic Doppler Shift
Suppose now that $\Delta t_{\rm obs}$ corresponds to the time between the arrival of two subsequent light wave crests. Then

\begin{align}
    \Delta t_{\rm obs} & = \frac{1}{\nu}, \: \Delta t' = \frac{1}{\nu'} \\
    \frac{1}{\nu} &= \frac{\gamma}{\nu'} \left(1 - \frac{v}{c} \cos \theta \right)\\
    \nu &= \frac{\nu'}{\gamma \left(1 - \frac{v}{c} \cos \theta \right)}\\
\end{align}
which is the relativistic Doppler shift. When the light source moves directly towards ($\theta = 0$) or away ($\theta = 180$) from the observer, then:

\begin{align}
    v_r &\equiv -v \cos \theta  = \mp v\\
    v_r^2 &= v^2
\end{align}
where $v_r$ is the radial component of the velocity, and the minus arises as we have said the source is moving towards us with a velocity $v$, but when we measure radial velocities from Earth, we typically measure a positive radial velocity as an object moving away from us, and a negative radial velocity as an object moving towards us.
We then get 
\begin{align}
    \gamma \left(1 + \frac{v_r}{c}\right) &= \frac{\left(1 + \frac{v_r}{c}\right)}{\sqrt{1 - \frac{v_r^2}{c^2}}}\\
    &= \frac{\left(1 + \frac{v_r}{c}\right)}{\sqrt{\left(1 + \frac{v_r}{c}\right)\left(1 - \frac{v_r}{c}\right)}}\\
    &= \sqrt{\frac{\left(1 + \frac{v_r}{c}\right)}{\left(1 - \frac{v_r}{c}\right)}}\\
\end{align}
and thus
$$
    \nu = \nu'\sqrt{\frac{\left(1 - \frac{v_r}{c}\right)}{\left(1 + \frac{v_r}{c}\right)}}
$$
where here $v_r$ is positive when source moving away from us, and negative when moving towards us. We can also write this in terms of wavelength using $\lambda \nu = c$:
$$
    \lambda = \lambda'\sqrt{\frac{\left(1 + \frac{v_r}{c}\right)}{\left(1 - \frac{v_r}{c}\right)}}
$$
- Motion towards the observer: higher $\nu$, shorter $\lambda$.
- Motion away the observer: lower $\nu$, longer $\lambda$.

## Z, reshift
The fractional shift in wavelength is called the redshift:
$$
    Z = \frac{\Delta \lambda}{\lambda'} = \frac{\lambda-\lambda'}{\lambda'}
$$
where $\lambda'$ is the wavelength in the rest frame. With the above expression for $\lambda$, we get 
$$
    Z = \sqrt{\frac{\left(1 + \frac{v_r}{c}\right)}{\left(1 - \frac{v_r}{c}\right)}}-1
$$

<div class="alert alert-block alert-info">
    <b>Example</b>

The emission lines in the spectrum of the quasar 3C273 display a redshift of $Z=0.158$ At what wavelength is the H$\beta$ line observed in the spectrum of this quasar?
</div>

The rest wavelength of H$\beta$ is 4861Å. So,
$$
    Z = \frac{\lambda-\lambda'}{\lambda'}
$$
Solving for $\lambda$ gives:
\begin{align}
    \lambda &= \lambda' (1+Z)\\
    &= 5636 Å
\end{align}

The spee with which the quasar is moving from the Earth (dur to the expansion of the Universe) is given by:
\begin{align}
    (Z+1)^2 &= \frac{1 + \frac{v_r}{c}}{1 - \frac{v_r}{c}}\\
    \frac{v_r}{c} &= \frac{(Z+1)^2-1}{(Z+1)^2+1} \\
    &=0.1456
\end{align}

## Sanity check
When $v_r/c<<1$, then we get (by using $(1+x)^n \sim 1 + nx$)

\begin{align}
    \left(1 + \frac{v_r}{c}\right)^{1/2}\left(1 - \frac{v_r}{c}\right)^{-1/2} & \sim \left(1+\frac{v_r}{2c}\right)\left(1+\frac{v_r}{2c}\right)\\
    & \sim 1+\frac{v_r}{c} + \frac{v_r^2}{4c^2}\\
    & \sim 1+\frac{v_r}{c}
\end{align}
Subbing this into
$$
    \lambda = \lambda'\sqrt{\frac{\left(1 + \frac{v_r}{c}\right)}{\left(1 - \frac{v_r}{c}\right)}}
$$
we get
$$
    \lambda = \lambda' \left( 1+\frac{v_r}{c} \right)
$$
which is the non relativistic Doppler shift (phew!).
