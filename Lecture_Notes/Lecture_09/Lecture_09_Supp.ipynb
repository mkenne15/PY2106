# Black-body radiation
## The partition function

Recall that at the end of the thermal physics course, we began to consider the partition function of an ideal quantum gas. As a recap, consider a gas of $N$ particles, which are non-interacting. For a perfect quantal gas, the energy of the system will be given by
$$
    U = \sum_{r} n_r \epsilon_r
$$
where this is a sum over all available states, and $n_r$ is the occupation number of each state. The occupation numbers are given by
\begin{align}
    n_r &= 0,1 & {\rm all \: r \: Fermi-Dirac}\\
    n_r &= 0,1,2,\dots & {\rm all \: r \: Bose-Einstein}
\end{align}

So, the partition function is then
\begin{align}
    Z(T,V,N) &= \sum_{n_1n_2,\dots,n_r} e^{-\beta \sum_{r} n_r \epsilon_r}\\
    Z(T,V,N) &= \sum_{n_1n_2,\dots,n_r} e^{-\beta (n_1 \epsilon_1+n_2 \epsilon_2+\dots)}\\
    Z(T,V,N) &= \sum_{n_1=0}^\infty e^{-\beta (n_1 \epsilon_1)} \sum_{n_2=0}^\infty e^{-\beta (n_2 \epsilon_2)} \dots \\
    Z(T,V,N) &= \prod_i \sum_{n_i=0}^\infty e^{-\beta (n_i \epsilon_i)}
\end{align}
That is, the partition function is the product over all states ($i$) of the sum of all occupation numbers($n_i$) of each state.
If we differentiate $\ln Z$ with respect to the energy level $\epsilon_i$ while leaving the temperature and other energy levels constant, we obtain the mean occupation number $\bar{n}_i$ of the state $i$. That is
$$
    \bar{n}_i = \frac{1}{\beta}\left(\frac{\partial \ln Z}{\partial \epsilon_i}\right)_{T,\epsilon_r(r!=i)}
$$

From this, we can calculate all desired quantities of the system at hand.

Borrowing a few more results from quantum mechanics, we will start with the following information.

- Photons have a spin of 1, meaning they act like bosons.
- Photons do not interact with each other, meaning a gas of photons acts like a perfect quantal gas

So if we now have a sealed container filled with photons, where the walls are constantly absorbing and emitting photons, what does the flux density within that container look like? The partition function is given by
\begin{align}
    Z &= \sum_{n_1=0}^\infty \sum_{n_2=0}^\infty \dots e^{-\beta (n_1 \epsilon_1+n_2 \epsilon_2+\dots)}\\
    Z &= \prod_i \sum_{n_i=0}^\infty e^{-\beta (n_i \epsilon_i)}
\end{align}
For bosons, where ${n_i} = {0,1,2,3,\dots}$, then this becomes
$$
    Z = \prod_i \left( 1 + e^{- \beta \epsilon_i} + e^{- 2\beta \epsilon_i}+\dots\right)
$$
which, using the geometric series $1+x+x^2+\dots = 1/(1-x)$ for $x<1$, gives
$$
    Z = \prod_i \frac{1}{1-e^{-\beta \epsilon_i}}
$$
Thus, we get
$$
    \ln (Z) = \sum_i \ln(1-e^{-\beta \epsilon_i})
$$
The average occupation number of each state is thus given by
$$
    \bar{n}_i = \frac{1}{\beta}\left(\frac{\partial \ln Z}{\partial \epsilon_i}\right) = \frac{1}{e^{\beta\epsilon_i}-1}
$$

The next quantity that we need is the density of states. Again, this comes from PY2104 (Lecture 12), and is given by
$$
    g(k){\rm d}k = \frac{1}{8}\frac{4\pi k^2 {\rm d} k}{(\pi/L)^3} 2 = \frac{V k^2 {\rm d} k}{\pi^2}
$$
where ${\bf k}=\frac{{\bf p}}{\hbar}$ is the wavevector. There is an extra factor of 2 in this term relative to what we dealt with in thermodynamics, and that is because electromagnetic waves can have 2 polarisations. Rewriting this in terms of momentum gives
$$
    g(p){\rm d}p = \frac{V p^2 {\rm d} p}{\pi^2 \hbar^3}
$$
Now, we know that the energy of a photon is $E=\hbar \omega$. As such, it is more convenient to rewrite the above expression in terms of $\omega$. The relation between $p$ and $\omega$ is just $p = \hbar \omega /c$, and so
$$
    g(\omega){\rm d}\omega = \frac{V \omega^2 {\rm d} \omega}{\pi^2 c^3}
$$
So, the mean number of photons between a frequency of $\omega$ and $\omega+{\rm d} \omega$ is then given by
$$
    {\rm d } N = \bar{n} g(\omega) {\rm d} \omega.
$$
Substituting in the above gives
$$
    {\rm d } N = \frac{V }{\pi^2 c^3} \frac{\omega^2 {\rm d} \omega}{e^{\beta \hbar \omega}-1}
$$
The energy contained by these photons is thus
$$
   U = \hbar \omega {\rm d } N = \frac{V \hbar}{\pi^2 c^3} \frac{\omega^3 {\rm d} \omega}{e^{\beta \hbar \omega}-1}
$$
The energy density ($u=U/V$) is then
$$
   u(\omega, T) {\rm d} \omega = \frac{ \hbar}{\pi^2 c^3} \frac{\omega^3 {\rm d} \omega}{e^{\beta \hbar \omega}-1}
$$
