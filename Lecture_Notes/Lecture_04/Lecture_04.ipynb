# The Virial Theorem

We've seen that gravity is well behaved and easy to deal with when we are only considering 2 objects, and that analytic expressions exist for the separation between the bodies $r(\theta)$, and both the kinetic and potential energies of the bodies. However, the addition of another body or bodies complicates things significantly. As such, when studying many body systems (galaxies, star clusters, planetary rings), we need statistical tools which can give us physical insight into the systems.

In todays lecture, we are going to consider one of these tools - the **virial theorem**. 


## Derivation
To start with, let's assume we have $N$ bodies, each with a mass $m_i$ and located at $\textbf{r}_i$. Now consider the function
$$
    A \equiv \sum^N_i m_i \frac{ {\rm d}\textbf{r}_i}{{\rm d} t} \cdot \textbf{r}_i.
$$
The time derivative of this function is
$$
    \frac{ {\rm d}A}{{\rm d} t} = \sum^N_i \left( m_i \frac{ {\rm d^2}\textbf{r}_i}{{\rm d} t^2} \cdot \textbf{r}_i + m_i \frac{ {\rm d}\textbf{r}_i}{{\rm d} t} \cdot \frac{ {\rm d}\textbf{r}_i}{{\rm d} t}\right)
$$
The second term is $2 K$, where $K$ is the kinetic energy. As for the first term, we can substitute one of the terms using Newton's second law:
$$
    \textbf{F}_i = m_i \frac{ {\rm d^2}\textbf{r}_i}{{\rm d} t^2}
$$
giving
$$
    \frac{ {\rm d}A}{{\rm d} t} = \sum^N_i \textbf{F}_i\cdot \textbf{r}_i + 2 K
$$
The left hand side can also be written as
\begin{align}
    \frac{{\rm d} A}{{\rm d} t} &= \frac{{\rm d} }{{\rm d} t} \sum^N_i m_i \frac{ {\rm d}\textbf{r}_i}{{\rm d} t} \cdot \textbf{r}_i\\
    &= \frac{{\rm d} }{{\rm d} t} \sum^N_i \frac{1}{2} \frac{ {\rm d}}{{\rm d} t} (m_i r_i^2)\\
    &= \frac{1}{2} \frac{{\rm d^2 I} }{{\rm d} t^2} \\
\end{align}
where $I$ is the moment of inertia of a point mass.
So our original expression can now be written as 
$$
  \frac{1}{2} \frac{{\rm d^2 I} }{{\rm d} t^2} = 2 <K> + \sum^N_i \textbf{F}_i\cdot \textbf{r}_i.
$$
This is known as the virial thereom. In order to make the equation a bit more comprehendable, let's take the time average of this equation
$$
    2 <K> + <\sum^N_i \textbf{F}_i\cdot \textbf{r}_i> = \frac{1}{2}<\frac{ {\rm d^2}I}{{\rm d} t^2} >
$$
Recalling that the average of a function over the time interval $0\to\tau$ can be written as (see multivariable calculus)
$$
    <\frac{ {\rm d^2}I}{{\rm d} t^2} > = \frac{1}{\tau} \int^\tau_0 \frac{{\rm d^2}I}{{\rm d} t^2} {\rm d} \tau
$$
we get
$$
    <\frac{ {\rm d^2}I}{{\rm d} t^2} > = \frac{1}{\tau} \left( \frac{{\rm d}I}{{\rm d} t} \bigg| _\tau - \frac{{\rm d}I}{{\rm d} t} \bigg|_0 \right)
$$
For a periodic system, then
$$
     \frac{{\rm d}I}{{\rm d} t} \bigg| _\tau =  \frac{{\rm d}I}{{\rm d} t} \bigg| _0
$$
and so the average over one period is 0. Infact, this is true for non-periodic functions too as $t \to \infty$, as long as $\frac{{\rm d}I}{{\rm d} t}$ is bounded. So, we end up with
$$
    2 <K> + <\sum^N_i \textbf{F}_i\cdot \textbf{r}_i> = 0
$$
which is a powerful tool, as it's true for any bounded system! Now, let's think about some astrophysical applications.

## In a gravitational field
Consider a single particle in our ensemble, and where the force acting on the particle is gravity. The force felt by this particle, $i$, due to every other particle is given by
$$
   \textbf{F}_i =  \sum_{j, j \neq i} \frac{G m_i m_j (\textbf{r}_j-\textbf{r}_i)}{|(\textbf{r}_j-\textbf{r}_i)^3|}
$$
Using this, it can be shown that
$$
    \sum^N_i \textbf{F}_i\cdot \textbf{r}_i = U
$$
which means the virial theorem becomes
$$
     2 <K> + <U> = 0
$$
or, remembering that $E = K + U$
$$
     2 <E> - <U> = 0
$$
So what use is this? To see, consider a gas cloud which is collapsing to form a star.

### Virial theorem example: Planet orbit
Consider a planet of mass $m$ moving in a circular orbit of radius $r$ around a star of mass $M$, such that $m\ll M$ and the star does not move (that is, it's located at the centre of mass). The potential energy of the system is then
$$
    U=âˆ’\frac{GMm}{r}
$$
The speed of the planet is given by equating the centrifugal force with the gravitational force felt by $m$:
$$
    \frac{mv^2}{r} = \frac{GMm}{r^2}
$$
which gives a kinetic energy of
$$
    K = \frac{GMm}{2r}
$$

Thus, we easily see in this case that $2<K>+<U>=0$.

### Virial theorem example: cloud collapse
To start with, let's assume we have a spherical cloud of uniform density $\rho$, mass $M$, and which is at an initial radius $R$. If we start with considering what the gravitational potential energy of a shell of this cloud is, we get
\begin{align}
    {\rm d}U &= - \frac{GM(r) {\rm d} m}{r}\\
    {\rm d}U &= - \frac{G(4/3 \pi r^3 \rho)(4 \pi r^2 \rho {\rm d} r)}{r}
\end{align}
where those substitutions come from considering the mass of the shell ${\rm d} m$, and the mass contained within the volume enclosed by the shell $M(r)$. The total potential is then
\begin{align}
    U &= - \int^R_0 \frac{G(4/3 \pi r^3 \rho)(4 \pi r^2 \rho {\rm d} r)}{r}\\
      &= - \frac{16}{3} \pi^2 \rho^2 G \int^R_0 r^4{\rm d} r\\
      &= - \frac{16}{3} \pi^2 \rho^2 G R^5\\
      &= - \frac{3 G M^2 }{5 R}
\end{align}

So let's now consider what happened with the Sun formed. Assuming we had a cloud of mass equal to that of the Sun $2\times10^30$ kg, but with an initial radius that's much larger than the Sun, that starts in equilibrium and ends in equilibrium, then the initial total amount of energy contained in the cloud was
$$
    E_i = \frac{1}{2}U_i = - \frac{3}{10} \frac{G M_{\odot}^2 }{R_i}
$$
After collapse, we have
$$
    E_{\odot} = - \frac{3}{10} \frac{G M_{\odot}^2 }{R_{\odot}}
$$
The change in energy is then
\begin{align}
    \Delta(E) &= - \frac{3}{10} \frac{G M_{\odot}^2 }{R_{\odot}} + \frac{3}{10} \frac{G M_{\odot}^2 }{R_i}\\
    \Delta(E) &\approx - \frac{3}{10} \frac{G M_{\odot}^2 }{R_{\odot}}\\
\end{align}
where we have assumed that the initial radius of the cloud was much much larger than the current radius of the Sun. This means that the change in energy is
$$
    \Delta(E) = -1.15\times10^41 \: {\rm J}.
$$
This is a huge amount of energy for the cloud to have lost!

### Virial theorem example: mass of galaxy cluster
Imagine we have a galaxy which has N stars in in. The kinetic energy contained within the galaxy is
$$
    K = \sum^N_{i=0} \frac{1}{2} m_i v^2_i = \frac{1}{2} M < v^2 >
$$
where $M$ is the total mass of the galaxy and $<v>$ mass-weighted mean square velocity of the stars, given by
$$
    <v^2> = \frac{1}{N} \sum^N_{i=1} v^2_{\rm i}
$$
The potential energy of this system is
$$
    U \sim -\frac{3}{5}\frac{G M^2}{R}
$$
which, using the virial theorem, means
$$
    M < v^2 > = \frac{3}{5}\frac{G M^2}{R}
$$
In order to proceed, we need an expression for $< v^2 >$. For the moment, let's assume the velocity distribution is isotropic, such that in spherical coordinates, we have
$$
    <v_{r}^2> =  <v_{\theta}^2> = <v_{\phi}^2>.
$$
This means that
$$
    <v^2> = 3 <v_{r}^2>.
$$
If the mean speed is zero, then the variance in $v_r$ is 
$$
    \sigma^2_r = \frac{1}{N} \sum^N_{i=1} v_{\rm r, i}^2 = <v_r>^2
$$
finally giving
$$
    <v^2> = 3 \sigma^2_r.
$$
The virial theorem then gives us that
$$
    M = \frac{5 R \sigma^2_r}{G}
$$
Consider now the Andromeda Galaxy. A velocity dispersion of 240 km s$^{-1}$ has been recorded within the inner 750pc. This equates to a virial mass of
$$
    M = \frac{5 R \sigma^2_r}{G} = 25\times10^6 \: {\rm M_{\odot}}
$$
