# Relativistic Velocity Transformations

In this lecture, we are going to study how velocities transform when using Einstein's postulates of special relativity. We start with using the Lorentz transformations to write the differential relations:
\begin{align}
    dx &= \frac{dx' + udt'}{\sqrt{1-u^2/c^2}}\\
    dy &= dy'\\
    dz &= dz'\\
    dt &= \frac{dt'+dx'u/c^2}{\sqrt{1-u^2/c^2}}.
\end{align}
We can use these to get the velocity transformations. For example:
$$
    v_x = \frac{dx}{dt} = \frac{dx' + udt'}{dt'+dx'u/c^2}
$$
If we then divide the top and bottom by $dt'$, while recognising that $\frac{dx'}{dt'}=v_x'$, then we get
\begin{align}
    v_x &= \frac{v_x' + u}{1+v_x'u/c^2}\\
    v_y &= \frac{v_y'}{\left(1+v_x'u/c^2\right) \gamma}\\
    v_z &= \frac{v_z'}{\left(1+v_x'u/c^2\right) \gamma}\\.
\end{align}
Here, $v_x'$, $v_y'$,$v_z'$ are the components of the velocity $\textbf{v'}$ of an object in the S' frame, and $v_x$, $v_y$,$v_z$ are the components of the velocity $\textbf{v}$ in the S frame. So, S' moves with a velocity $\textbf{u}$ with the respect to S in the positive x direction.

$v_y$ and $v_z$ transform in the same way because they are both perpendicular to the velocity $\textbf{u}$ between S and S'. We can write these in a more compact form by considering just 2 components of $\textbf{v}$ - one which is parallel to $\textbf{u}$ and one which is perpendicular to $\textbf{u}$:
\begin{align}
    v_\parallel &= \frac{v_\parallel' + u}{1+v_\parallel'u/c^2}\\
    v_\perp &= \frac{v_\perp'}{\gamma \left(1+v_\parallel'u/c^2\right)}\\
\end{align}
Find the relationship between angles along which an object moves:
$$
   \tan \phi = \frac{v_\perp}{v_\parallel} = \frac{v_\perp'}{\gamma \left( v_\parallel' + u \right)}\\
$$
Now if we note that $v_\perp'=v'\sin \phi'$ and $v_\parallel'=v'\cos \phi'$
$$
   \tan \phi = \frac{v'\sin \phi'}{\gamma \left( v'\cos \phi' + u \right)}\\
$$
$\phi'$ is the direction of motion in $S'$, while $\phi$ is the direction of motion in $S$, all relative to $u$.

So, how do we use and interpret these expressions? What does this apparent change in the direction mean? Consider a photon, which means that $v'=c$. This gives:
$$
   \tan \phi = \frac{\sin \phi'}{\gamma \left( \cos \phi' + u/c \right)}\\
$$
Now, consider a photon emitted at $90^{\rm o}$ to the velocity $\textbf{u}$ in the S' frame. For this setup, $\sin \phi'=1$ and $\cos \phi'=0$. And so
\begin{align}
   \tan \phi &= \frac{1}{\gamma u/c} = \frac{1}{\gamma \beta}\\
   \cos \phi &= \frac{1}{\sqrt{\tan^2 \phi +1}} = \beta\\
   \sin \phi &= \sqrt{1-\cos^2 \phi} = \frac{1}{\gamma}
\end{align}
This tells us what the apparent direction of motion a photon will have for an observer in the S frame, assuming the photon was emitted at an angle of $90^{\rm o}$ in the S' frame. So, since $\gamma>1$, then $\frac{1}{\gamma}<1$. This means:
\begin{align}
    \sin \phi &= \frac{1}{\gamma}\\
    \sin \phi &< 1\\
    \phi &< 90^{\rm o}
\end{align}
The photon is observed in the observer's (S) frame to travel at a smaller angle to the direction of source motion.
A source's radiation (the photons it emits) is concentrated in the forward direction of its motion. Independant of $\beta$, all photons with $0^{\rm o} <\phi'<90^{\rm o}$ will be concentrated in a smaller angle in the S frame.

## Example: A source which is moving with $\beta = u/c = 0.5$

Consider photons emitted at specific angles in the S' frame, say $\phi'=45^{\rm o}, 90^{\rm o}, 120^{\rm o}, \: {\rm and} \: 135^{\rm o}$. In each case, 
$$
    \cos \phi = \frac{\cos \phi'+0.5}{1+0.5\cos \phi'}.
$$
This yields the following values for $\cos \phi$ and $\phi$:

| $\phi'$ | $\cos \phi'$ | $\cos \phi$ | $\phi$ |
| --- | --- | --- | --- |
| $45^{\rm o}$ | $1/\sqrt{2}$ | $0.89$ | $27^{\rm o}$ |
| $90^{\rm o}$ | $0$ | $1/2$ | $60^{\rm o}$ |
| $120^{\rm o}$ | $1/2$ | $0$ | $90^{\rm o}$ |
| $135^{\rm o}$ | $-1/\sqrt{2}$ | $-0.32$ | $109^{\rm o}$ |

All photons emitted within $120^{\rm o}$ of direction of motion in S' will be observed in frame S to travel in the forward hemisphere!

In general, $\phi=90^{\rm o}$, then $\cos \phi = 0$, which is true when:
$$
    \cos \phi'=-\beta
$$
This is the condition for photons to be observed in the forward hemisphere in frame S.
For example: 

| $\beta$ | 0.5 | 0.7 | 0.9 | 0.95 | 0.99 |
| --- | --- | --- | --- | --- | --- |
| $\phi '$ | 120 | 134 | 154 | 162 | 172 |

This is called the **headlight effect** or **Doppler Boosting**.

One consequence of this is that a moving source of radiation will appear brighter than it is in its rest frame when it approaches an observer, and less bright than its rest frame when it receeds from an observer.

We see the Doppler boost effect in the pulses from pulsars only when the pulsar beam is roughly sligned with the Earth. Below shows an example of how a photons angle is changed by this transformation.

```python
import numpy as np
import matplotlib.pyplot as plt

beta = 0.5
phi_p = np.linspace(0,2*np.pi,100)
r = np.ones_like(phi_p)

cos_phi = (np.cos(phi_p)+beta)/(1+0.5*np.cos(phi_p))
phi = np.arccos(cos_phi)
phi[phi_p>np.pi] = np.pi+(np.pi-phi[phi_p>np.pi])

fig = plt.figure(figsize=[8,4], dpi=150)
ax = fig.add_subplot(121,projection='polar')
ax.set_title("In the S' frame, the light is uniformly emitted.")
ax.scatter(phi_p, r, c=-np.cos(phi_p), s=20.0, cmap='bwr')

ax = fig.add_subplot(122,projection='polar')
ax.set_title(r"In the S frame, the light is beamed in the"+"\n"+"same direction as motion as the source." + "\n" + "Note that the direction of motion is to the right.")
ax.scatter(phi, r, c=-np.cos(phi_p), s=20.0, cmap='bwr')
plt.tight_layout()
plt.savefig("Figures/Dopp_beaming.png")
plt.show()
```

In the above, the points are coloured by the value of the angle in the unprimed frame - that is, the white points in the primed frame get transferred in position to where the white points are in the unprimed frame. There are two things we can draw from the above plot. First, look at the faint dots close to $90^{\rm o}$ and $270^{\rm o}$ in the S' frame. The photons emitted at these angles in this frame get transformed to around to $60^{\rm o}$ and $310^{\rm o}$. Even more interestingly, some of the red dots which are emitted opposite the direction of motion in S' end up moving towards the forward hemisphere, meaning an observer located at $0^{\rm o}$  in S (such that the object is travelling towards them) can see around the back of the object!

Additionally, look at the density of points in the S frame. The density is much higher in the hemisphere on the right - meaning more light is being emitted in this direction, and less in the direction opposite the motion. This means an observer located at $0^{\rm o}$  in S (such that the object is travelling towards them) would see the object as being brighter than the object is in its rest frame, while an observer located at $180^{\rm o}$  in S (such that the object is travelling away from them) would see the object as being fainter than it is in its rest frame.
