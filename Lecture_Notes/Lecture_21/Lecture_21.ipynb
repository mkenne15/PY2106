# Relativistic Doppler Shift

In this lecture, we're going to discuss how the concepts of special relativity affect the Doppler shift, which was introduced previously in this course. We'll start by considering general time intervals, then reconsider this in the contact of the frequency and wavelength of light. Finally, we'll quickly show that what we've derived reduces to the expected clssical Doppler shift for low velocities.

## The Doppler Factor
Consider some light source that is moving at an angle to our line of sight. We're going to set up our frames such that we are in the **S** frame, and that **S'** moves in the same direction and with the same speed as the light source, such that the source is at rest in the primed frame. This source emits two photons, at time $t_1$ and $t_2$. I've sketched this setup below. Note that this differs a little bit from what I did in the lectures, but I think this is easier to follow

![Rel_Dop_Shift](Figures/Rel_Doppler_Shift.png)

In the time interval between the emission of the photons in our frame, the source will have moved a distance of $u\Delta t$. Since the photons are emitted at the same x' coordinate in the frame of the source, we only have to worry about time dialation for this conversion, meaning the source has moved a distance of $\gamma u\Delta t'$, where $t'$ is the time between the emission of the two photons as measured in the frame of the source. Thus, photon two has an extra bit of distance it needs to cover before it reaches us compared to photon one, given by $\gamma u\Delta t' \cos(\theta)$. The time taken to cover this extra distance is $\gamma \frac{u}{c}\Delta t' \cos(\theta)$. Thus, the time interval between both photons arriving at Earth is given by the time dialation moving between both frames and the extra time taken by photon 2 to travel this additional distance - that is
$$
    \Delta t_{\rm obs} = \gamma \Delta t' + \gamma \frac{u}{c}\Delta t' \cos(\theta) = \gamma \Delta t' \left( 1 + \frac{u}{c} \cos \theta \right )
$$

There's an interesting consequence to this equation. Consider a source that is moving at right angles to the line of sight ($\theta=90^{\rm o}$). Even though the source has no velocity component towards us, there is still a shift due to time delay - this is a unique prediction of special relativity over the classical Doppler effect.

Note that, in many places, you'll see the equation above written as

$$
    \Delta t_{\rm obs} = \gamma \Delta t' \left( 1 - \frac{u}{c} \cos \phi \right )
$$
The change in sign here occurs because other derivations use the angle $\phi$ rather than $\theta$ - but of course, $\cos(\phi)=-\cos(\theta)$ since $\phi+\theta=180^{\rm o}$, so everything is still the same. Just be sure which angle you are measuring when using these equations.

## Time interval durations
At this stage, it's worth asking ourselves the question: Is the observed time interval shorter or longer than the time interval in the rest frame of the source ($\Delta t'=\Delta t_{\rm rest}$)?
Starting with:
$$
    \Delta t_{\rm obs} = \gamma \Delta t_{\rm rest} \left(1 + \frac{u}{c} \cos \theta \right)
$$
it's not clear there is a simple answer, because of the $\cos$ term. Let's break down what we know and see if we can get some insight

### Source moving away from us ($-90 < \theta \leq 90$)

- $\gamma > 1$ always.
- $\left(1 + \frac{u}{c} \cos \theta \right) >1$ when $-90 < \theta \leq 90$

Thus the two factors act in the same sense when the source is moving away from us (that is, they are both $>1$). This means that
$$
    \Delta t_{\rm obs} > \Delta t_{\rm rest}.
$$
This means the time interval in the rest frame is shorter than observed. 

### Source moving nearly directly towards us ($\theta \sim 180$ )

Now, when $\theta \sim 180$ such that $\cos \theta \sim -1$, then
$$
    \gamma \left(1 + \frac{u}{c} \cos \theta \right) < 1
$$
the observed time interval will be shorter than the rest frame interval
$$
    \Delta t_{\rm obs} < \Delta t_{\rm rest}.
$$
This can be very useful when you observe an event and know the sources motion relative to you, as it can allow you to figure out how long it took in the rest frame of the source. This becomes important for objects which may be moving towards us at very hight velocities, as events which seem to take seconds in our frame (and hence are hard to physically explain) may take much longer in the rest frame of the source.

## Moving to frequency and wavelength
Suppose now that $\Delta t_{\rm obs}$ corresponds to the time between the arrival of two subsequent light wave crests. Then

\begin{align}
    \Delta t_{\rm obs} & = \frac{1}{\nu}, \: \Delta t' = \frac{1}{\nu'} \\
    \frac{1}{\nu} &= \frac{\gamma}{\nu'} \left(1 + \frac{u}{c} \cos \theta \right)\\
    \nu &= \frac{\nu'}{\gamma \left(1 + \frac{u}{c} \cos \theta \right)}\\
\end{align}
which is the relativistic Doppler shift. When the light source moves directly away ($\theta = 0$) or towards ($\theta = 180$) from the observer, then:

\begin{align}
    v_r &\equiv u \cos \theta  = \pm u\\
    v_r^2 &= v^2
\end{align}
where $v_r$ is the radial component of the velocity.
We then get 
\begin{align}
    \gamma \left(1 + \frac{v_r}{c}\right) &= \frac{\left(1 + \frac{v_r}{c}\right)}{\sqrt{1 - \frac{v_r^2}{c^2}}}\\
    &= \frac{\left(1 + \frac{v_r}{c}\right)}{\sqrt{\left(1 + \frac{v_r}{c}\right)\left(1 - \frac{v_r}{c}\right)}}\\
    &= \sqrt{\frac{\left(1 + \frac{v_r}{c}\right)}{\left(1 - \frac{v_r}{c}\right)}}\\
\end{align}
and thus
$$
    \nu = \nu'\sqrt{\frac{\left(1 - \frac{v_r}{c}\right)}{\left(1 + \frac{v_r}{c}\right)}}
$$
where here $v_r$ is positive when source moving away from us, and negative when moving towards us. We can also write this in terms of wavelength using $\lambda \nu = c$:
$$
    \lambda = \lambda'\sqrt{\frac{\left(1 + \frac{v_r}{c}\right)}{\left(1 - \frac{v_r}{c}\right)}}
$$
- Motion towards the observer: higher $\nu$, shorter $\lambda$.
- Motion away the observer: lower $\nu$, longer $\lambda$.

## Z, redshift
The fractional shift in wavelength is called the redshift:
$$
    Z = \frac{\Delta \lambda}{\lambda'} = \frac{\lambda-\lambda'}{\lambda'}
$$
where $\lambda'$ is the wavelength in the rest frame. With the above expression for $\lambda$, we get 
$$
    Z = \sqrt{\frac{\left(1 + \frac{v_r}{c}\right)}{\left(1 - \frac{v_r}{c}\right)}}-1
$$

<div class="alert alert-block alert-info">
    <b>Example</b>

The emission lines in the spectrum of the quasar 3C273 display a redshift of $Z=0.158$ At what wavelength is the H$\beta$ line observed in the spectrum of this quasar?
</div>

The rest wavelength of H$\beta$ is 4861Å. So,
$$
    Z = \frac{\lambda-\lambda'}{\lambda'}
$$
Solving for $\lambda$ gives:
\begin{align}
    \lambda &= \lambda' (1+Z)\\
    &= 5629 Å
\end{align}

The speed with which the quasar is moving from the Earth (due to the expansion of the Universe) is given by:
\begin{align}
    (Z+1)^2 &= \frac{1 + \frac{v_r}{c}}{1 - \frac{v_r}{c}}\\
    \frac{v_r}{c} &= \frac{(Z+1)^2-1}{(Z+1)^2+1} \\
    &=0.1456
\end{align}

## Sanity check
When $v_r/c<<1$, then we get (by using $(1+x)^n \sim 1 + nx$)

\begin{align}
    \left(1 + \frac{v_r}{c}\right)^{1/2}\left(1 - \frac{v_r}{c}\right)^{-1/2} & \sim \left(1+\frac{v_r}{2c}\right)\left(1+\frac{v_r}{2c}\right)\\
    & \sim 1+\frac{v_r}{c} + \frac{v_r^2}{4c^2}\\
    & \sim 1+\frac{v_r}{c}
\end{align}
Subbing this into
$$
    \lambda = \lambda'\sqrt{\frac{\left(1 + \frac{v_r}{c}\right)}{\left(1 - \frac{v_r}{c}\right)}}
$$
we get
$$
    \lambda = \lambda' \left( 1+\frac{v_r}{c} \right)
$$
which is the non relativistic Doppler shift (phew!).
